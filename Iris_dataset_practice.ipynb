```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

#first lets import the dataset

iris=pd.read_csv("IRIS.csv")
```

```python
iris.head()
```

```python
# Creating a DataFrame. A
df = pd.DataFrame(iris)

# Display the first 5 rows of the DataFrame
df.head()
```

Now lets see the statistical properties of iris dataset


```python
#describing the statistical properties
df.describe()
```

So we get some idea about how the dataset looks. Lets see if there are any missing values here

```python
df.isnull()
```

```python
df.isnull().sum()
```

No missing values here. Now we calculate the mean median and mode of every attribute

```python
mean1 = df['sepal_length'].mean()
mean2 = df['sepal_width'].mean()
mean3 = df['petal_length'].mean()
mean4 = df['petal_width'].mean()
print(mean1)
print(mean2)
print(mean3)
print(mean4)
```

```python
#mean, median and mode, SD and range of all attributes
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']


for col in columns:
    mean_value = df[col].mean()
    median_value= df[col].median()
    mode_value= df[col].mode()
    std_value = df[col].std()
    range_value = df[col].max() - df[col].min()

    print(f"Mean of {col}: {mean_value}")
    print(f"Median of {col}: {median_value}")
    print(f"Mode of {col}: {mode_value}")
    print(f"Standard Deviation of {col}: {std_value}")
    print(f"Range of {col}: {range_value}")
    print("\n")
```

Now after this. We are doing Data Visualization

```python
# Histogram for each feature
df.hist(bins=20, figsize=(5, 7))
plt.show()

```

```python

for col in df.columns[:-1]: #we are skipping the species column
    sns.histplot(df[col], kde=True, color='orange')
    plt.show()
```

# Now lets create box plot and scatter plot to see if there is any outliers

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.boxplot(df[columns], vert=True)

plt.xticks([1, 2, 3, 4], columns)

plt.title('Boxplot of Iris Dataset Attributes')
plt.ylabel('Values')
plt.show()
```

```python
sns.pairplot(df, hue='species')
plt.show()
```

```python
attri_df = df.drop(['species'], axis=1)
# Calculate the correlation matrix
co_mat = attri_df.corr()

# Plot the heatmap
sns.heatmap(co_mat, annot=True, cmap='Purples')
plt.title('Heatmap of relationship of different attribute')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

x= df.drop(['species'], axis=1)
y= df['species']
# Split with stratification. as without this. the model was giving 100%accuracy
x_train, x_temp, y_train, y_temp = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
x_val, x_test, y_val, y_test = train_test_split(x_temp, y_temp, test_size=0.5, random_state=42, stratify=y_temp)
```

# We are using two simple models for prediction task
1. Naive Bayes
2. Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
#decision tree
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)




```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

y_pred_dt = dt.predict(x_val)
accuracy_dt = accuracy_score(y_val, y_pred_dt)
precision_dt = precision_score(y_val, y_pred_dt, average='macro')
recall_dt = recall_score(y_val, y_pred_dt, average='macro')
f1_dt = f1_score(y_val, y_pred_dt, average='macro')
print(f'Performance matrix of decision trees- Accuracy: {accuracy_dt}, Precision: {precision_dt}, Recall: {recall_dt}, F1 Score: {f1_dt}')
```

```python

from sklearn.naive_bayes import GaussianNB

# Naive Bayes
nb = GaussianNB()
nb.fit(x_train, y_train)




```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

y_pred_nb = nb.predict(x_val)
accuracy_nb = accuracy_score(y_val, y_pred_nb)
precision_nb = precision_score(y_val, y_pred_nb, average='macro')
recall_nb = recall_score(y_val, y_pred_nb, average='macro')
f1_nb = f1_score(y_val, y_pred_nb, average='macro')

print(f'Naive Bayes - Accuracy: {accuracy_nb}, Precision: {precision_nb}, Recall: {recall_nb}, F1 Score: {f1_nb}')
```

```python
# training accuracy
train_accuracy_dt = accuracy_score(y_train, dt.predict(x_train))
val_accuracy_dt = accuracy_score(y_val, dt.predict(x_val))

print(f"Decision Tree - Training Accuracy: {train_accuracy_dt}, Validation Accuracy: {val_accuracy_dt}")

train_accuracy_nb = accuracy_score(y_train, nb.predict(x_train))
val_accuracy_nb = accuracy_score(y_val, nb.predict(x_val))

print(f"Naive Bayes - Training Accuracy: {train_accuracy_nb}, Validation Accuracy: {val_accuracy_nb}")
```
